### Téléchargement des données de réanalyse climatique horaire de ERA5-land du ECMWF (European Centre for Medium-Range Weather Forecasts)


Pour réaliser une requête API, le ECMWF demande une clé, cette clé d'authenfication est gratuite et demande simplement la création d'un compte. Enssuite le code de la clé est consultable depuis le lien : https://cds.climate.copernicus.eu/api-how-to#install-the-cds-api-key

Il faut ensuite Copier/Coller le texte avec la clé personnelle ci dessous (ici la mienne) dans le fichier "USER/.cdsapirc" (à la base du répertoire utilisateur dans l'environnement Windows). 

Le fichier doit être créé à l'aide du Bloc-notes : "Fichier> Enregistrer sous> Type : Tous les fichiers> Nom du fichier : .cdsfapirc"

```python
# Chemin et fichier d'enregistement 
folder_out = r"/home/tolletg/notebooks/METEO FRANCE/Export data ERA-5"
downloaded_file = 'ERA5-Land hourly.nc'

# Emprise spatiale de la zone à extraire
lon_min, lat_min, lon_max, lat_max =[1.56274,44.458012,2.47075,44.88762] # x = coin inf gauche, y= coin inf gauche, x = coin sup droit y= coin sup 

# Choix de la période
start_year = 2024                                             
end_year = 2024                                             
month = ['02'] # ['01','02','03'] pour plusieurs mois

#Choix des paramètres
variables_list = 'total_precipitation' # voir liste des paramètres

# Téléchargement

downloaded_file = os.path.join(folder_out, downloaded_file)
years = [ str(start_year +i ) for i in range(end_year - start_year + 1)] 
start_day = 1
end_day = 31
days = [ str(start_day +i ).zfill(2) for i in range(end_day - start_day + 1)]

print('Process started. Please wait the ending message ... ')
start = datetime.datetime.now()

c = cdsapi.Client()

c.retrieve(
    'reanalysis-era5-land',
    {
        'year': years,
        'variable': variables_list,
        'month': month,
        'day': days,
        'time': [
            '00:00', '01:00', '02:00',
            '03:00', '04:00', '05:00',
            '06:00', '07:00', '08:00',
            '09:00', '10:00', '11:00',
            '12:00', '13:00', '14:00',
            '15:00', '16:00', '17:00',
            '18:00', '19:00', '20:00',
            '21:00', '22:00', '23:00',
        ],
        'area': [ lat_min, lon_min, lat_max, lon_max ],
        'format': 'netcdf',
    },
    downloaded_file)

print('Process completed in ', datetime.datetime.now()-start)
```

```python
# Calculer la somme cumulative des précipitations sur 24 heures pour chaque jour
daily_precipitation = df['tp'].rolling(time=24, center=False).sum()

# Sélectionner uniquement les valeurs de précipitations à 00:00 UTC pour chaque jour
daily_precipitation = daily_precipitation.sel(time=daily_precipitation['time.hour'] == 0)

# Convertir les précipitations de mètres en millimètres
daily_precipitation_mm = daily_precipitation * 1000

# Sélectionner la période spécifiée
daily_precipitation_period = daily_precipitation_mm.sel(time=slice("2024-02-01", "2024-02-29"))

# Calculer la somme des précipitations quotidiennes sur la période sélectionnée
total_precipitations_mm = daily_precipitation_period.sum(dim=('latitude', 'longitude'))

# Calcul du nombre total de mailles dans la grille
num_cells = df.sizes['latitude'] * df.sizes['longitude']

# Division des précipitations cumulées par le nombre de mailles pour obtenir des mm par maille
precipitation_values_per_cell = total_precipitations_mm.values / num_cells

# Récupérer les dates
dates = total_precipitations_mm.time.values

# Créer un graphique à barres
plt.figure(figsize=(10, 6))
plt.bar(dates, precipitation_values_per_cell)

# Ajouter des attributs
plt.title('Précipitations moyennes quotidienne ERA5-Land')
plt.xlabel("Date")
plt.ylabel("milimetre(mm)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Calcul du cumul total sur la période
total_cumul = np.sum(precipitation_values_per_cell)
print(f"Cumul total: {total_cumul:.2f} mm")
```
